```python
import pandas as pd
```

```python
incidents = pd.read_csv('NYPD_Use_of_Force_Incidents_20260415.csv')
members = pd.read_csv('NYPD_Use_of_Force__Members_of_Service_20260415.csv')
subjects = pd.read_csv('NYPD_Use_of_Force__Subjects_20260415.csv')
```

```python
display(incidents.head())
display(members.head())
display(subjects.head())
```

```python
# Get all incidents that have only one subject, create a row with subject info and number of officers involved
counts = subjects['TRI Incident Number'].value_counts()
single_subject_tri = counts[counts < 2]
single_subject = subjects.copy()
single_tris = single_subject_tri.index.to_list()
single_subject = single_subject[single_subject['TRI Incident Number'].isin(single_tris)]
members_single_subject = members[members['TRI Incident Number'].isin(single_tris)]
print(len(single_subject))
print(len(members_single_subject))
single_subject.sort_values(by='TRI Incident Number', inplace=True)
single_tris.sort()
member_counts = []

for tri in single_tris:
  member_count = len(members_single_subject[members_single_subject['TRI Incident Number'] == tri])
  member_counts.append(member_count)

single_subject['MemberCount'] = member_counts
single_subject_full = pd.merge(left=incidents, right=single_subject, on='TRI Incident Number', suffixes=['_x', None])
print(single_subject_full.head())
```

```python
single_subject_full.to_csv('TRI_SingleSubject_AnyMembers.csv', index=False)
```
